# **1st ML model**

```python
import pandas as pd
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/Patilanuj/Python_Projects_for_Data_Analysis/refs/heads/main/Delaney_solubility_with_discreptors/delaney_solubility_with_descriptors.csv')
df
```

## **Data Preparation**

```python
y=df['logS']
y
```

```python
x=df.drop('logS',axis=1)
x
```

## **Splitting X and Y for train and test**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=100)
```

```python
x_train
```

```python
x_test
```

## **Model building**

## **Linear regression**

####Training model

```python
#Training model
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

####Applying model to make preciction

```python
y_lr_train_pred=lr.predict(x_train)
y_lr_test_pred=lr.predict(x_test)
```

```python
y_lr_train_pred
```

```python
y_lr_test_pred
```

####Evaluate Model Performance

```python
from sklearn.metrics import mean_squared_error,r2_score
lr_train_mse=mean_squared_error(y_train,y_lr_train_pred)
lr_train_r2=r2_score(y_train,y_lr_train_pred)
lr_test_mse=mean_squared_error(y_test,y_lr_test_pred)
y_lr_test_r2 =r2_score(y_test,y_lr_test_pred)
```

```python
print("lr mse(train:)",lr_train_mse)
print("lr r2(train:)",lr_train_r2)
print("lr mse(test:)",lr_test_mse)
print("lr r2(test:)",y_lr_test_r2)
```

```python
lr_results = pd.DataFrame(['Linear Regression', lr_train_mse, lr_train_r2, lr_test_mse, y_lr_test_r2]).transpose()
```

```python
lr_results
```

```python
lr_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
```

```python
lr_results
```

## **Random Forest**

### Training model

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor(max_depth=2,random_state=100)
rf.fit(x_train,y_train)
```

### Applying model to make prediction

```python
y_rf_train_pred=rf.predict(x_train)
y_rf_test_pred=rf.predict(x_test)
```

### Evaluate mdoel performance

```python
from sklearn.metrics import mean_squared_error , r2_score
rf_train_mse=mean_squared_error(y_train,y_rf_train_pred)
rf_train_r2=r2_score(y_train,y_rf_train_pred)
rf_test_mse=mean_squared_error(y_test,y_rf_test_pred)
rf_test_r2=r2_score(y_test,y_rf_test_pred)
```

```python
print("rf mse(train:)",rf_train_mse)
print("rf r2(train:)",rf_train_r2)
print("rf mse(test:)",rf_test_mse)
print("rf r2(test:)",rf_test_r2)
```

```python
rf_results = pd.DataFrame(['Random Forest', rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2 ]).transpose()
rf_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
```

```python
rf_results
```

## **Model Comparison**

```python
df_models = pd.concat([lr_results,rf_results], axis=0).reset_index(drop = True)
df_models
```

## **Visualisation of the model results**

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figsize=(5,5)
plt.scatter(x=y_train, y=y_lr_train_pred,color ="#7CAE00", alpha=0.4)
z= np.polyfit(y_train,y_lr_train_pred,1)
p=np.poly1d(z)
plt.plot(y_train,p(y_train),"r--")
plt.xlabel("Experimental logS")
plt.ylabel("Predicted logS")
plt.show()
```
