# Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference. (5 points) 

```python
# converting values of 'GoalsFor' and 'GoalsAgainst' column into int datatype
var1['GoalsFor'] = var1['GoalsFor'].astype(int)
var1['GoalsAgainst'] = var1['GoalsAgainst'].astype(int)
```

```python
# defining function 'Goal_diff_count()' as per the problem statement needs
def Goal_diff_count():   
     var1['Goal_diff_count'] = var1['GoalsFor']-var1['GoalsAgainst']
     return var1[['Team','Goal_diff_count']]
Goal = Goal_diff_count()
Goal.sort_values(by = 'Goal_diff_count',ascending=False)
```

# Create a new column with name “Winning Percent” and append it to the data set (5 points)

Percentage of Winning = (GamesWon / GamesPlayed)*100

If there are any numerical error, replace it with 0%

Print the top 5 teams which has the highest Winning percentage



```python
var1['GamesWon'] = var1['GamesWon'].astype(int)
var1['GamesPlayed'] = var1['GamesPlayed'].astype(int)
```

```python
# calculate winning percentage for each team and storing value to new column 'Win Per'
var1['Win Per'] = (var1['GamesWon']/var1['GamesPlayed']) *100
var1['Win Per'].fillna(0,inplace = True)
var1[['Team','Win Per']]
```

```python
# Top 5 teams with highest winning percentage
var1[['Team','Win Per']].head(5)
```
